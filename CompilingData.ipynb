```python
import pandas as pd
import glob
```

```python
#Appending 24 districts data into one dataframe

data = pd.DataFrame()
for filename in glob.glob('JHARKHAND*'):
    df = pd.read_csv(filename)
    data = data.append(df)       
```

```python
len(data)
```

```python
data.to_csv('ALL.csv',index=False)
```

```python
data = pd.read_csv(r'ALL.csv')
data.head()
```

```python
l = []
m = []
n = []
for index, row in data.iterrows():
    #Extracting Gram Panchayat Number
    m.append(str(row['Gram Panchayat Name'])[-7:-1])
    
    #Making GP Code Concat : State Code + District Code + Block Code + Panchayat Code
    n.append(str('20'+str(row['District Name'])[-4:-1] + str(row['Block Name'])[-5:-1] + str(row['Gram Panchayat Name'])[-7:-1]))
    
    i = str(row['PDF']).find('Name=')
    if i!=-1:
        l.append(str(row['PDF'])[i+5:])
    else:
        l.append(None)
        
data['pdf_names'] = l
data['GP Code'] = m
data['GPCode_Concat'] = n
```

```python
data.to_csv('ALL_new.csv',index=False)
```
